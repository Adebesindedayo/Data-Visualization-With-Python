# Dataset on Immigration to Canada

```python
import numpy as  np   #  scientfic computing (ndarray library) 
import pandas as pd   #  data structure tool using to mainpulate and anysis dataset
```

```python
#exist mulitple functions in pandas can read data with multi format  (csv , xlsx , json , ..... )
df_can = pd.read_excel('Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skip_footer=2)
df_can.head() # head function display the first five rows in dataframe
```

# Dataset Preprocssing

```python
 # in pandas axis=0 represents rows (default) and axis=1 represents columns.

df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
 
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
 
df_can.columns = list(map(str, df_can.columns))
 
df_can.set_index('Country', inplace=True)   
 
df_can['Total'] = df_can.sum(axis=1) # add new column to sum number of immigraints for each counry
```

# Line Plot 

```python
years = list( map ( str , range(1980,2014) ) ) 
haiti = df_can.loc['Haiti', years] 
haiti.index = haiti.index.map(int) # Test this cell without/with this line to see the difference
haiti.plot(kind='line') 
plt.title('immigration form Haiti') #add title of plot
plt.xlabel('Years') #add xlabel text
plt.ylabel('Number of immigraint')  # add ylabel text
plt.text(2000, 6000, '2010 Earthquake')  # add text within plot at positions x = 2000 , y = 6000
plt.savefig('immigration from haiti.png') #to save the figure locally  in your machine
plt.show()
```

# compare the number of immigrants from India and China from 1980 to 2013.

```python
years = list( map ( str , range(1980,2014) ) ) 
df_new = df_can.loc[['India','China'], years] 
df_new = df_new.transpose()
df_new.index = df_new.index.map(int) # Test this cell without/with this line to see the difference
df_new.plot(kind='line' , figsize=(14,8)) 
plt.title('immigration form India and China') #add title of plot
plt.xlabel('Years') #add xlabel text
plt.ylabel('Number of immigraint')  # add ylabel text
plt.savefig('immigration form India and China.png') #to save the figure locally  in your machine
plt.show()
```

# compare the trend of top 5 countries that contributed the most to immigration to Canada. 

```python
years = list( map ( str , range(1980,2014) ) ) 

df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

df_top5.index = df_top5.index.map(int) # Test this cell without/with this line to see the difference [convert string year to int]

df_top5.plot(kind='line' ,figsize=(14,8)) 
plt.title('immigration form top 5 country ') #add title of plot
plt.xlabel('Years') #add xlabel text
plt.ylabel('Number of immigraint')  # add ylabel text
plt.savefig('immigration form 5 top country .png') #to save the figure locally  in your machine
plt.show()
```
